```python
%%bash

cd ~

git clone --recursive https://github.com/RSDKModding/RSDKv3-Decompilation.git

cd RSDKv3-Decompilation

git submodule update --init --recursive
```

```python
%%bash

set -e

cd ~/RSDKv3-Decompilation

echo "=== Installing Android dependencies ==="

mkdir -p dependencies/android
cd dependencies/android

# SDL2
echo "[1/5] Downloading SDL2..."
wget -q https://github.com/libsdl-org/SDL/releases/download/release-2.28.3/SDL2-2.28.3.tar.gz
tar -xf SDL2-2.28.3.tar.gz
mv SDL2-2.28.3 SDL
rm SDL2-2.28.3.tar.gz

# libogg
echo "[2/5] Downloading libogg..."
wget -q https://downloads.xiph.org/releases/ogg/libogg-1.3.5.tar.gz
tar -xf libogg-1.3.5.tar.gz
mv libogg-1.3.5 libogg
rm libogg-1.3.5.tar.gz

# libvorbis
echo "[3/5] Downloading libvorbis..."
wget -q https://downloads.xiph.org/releases/vorbis/libvorbis-1.3.7.tar.gz
tar -xf libvorbis-1.3.7.tar.gz
mv libvorbis-1.3.7 libvorbis
rm libvorbis-1.3.7.tar.gz

# libtheora
echo "[4/5] Downloading libtheora..."
wget -q https://downloads.xiph.org/releases/theora/libtheora-1.1.1.tar.gz
tar -xf libtheora-1.1.1.tar.gz
mv libtheora-1.1.1 libtheora
rm libtheora-1.1.1.tar.gz

# GLEW
echo "[5/5] Downloading GLEW..."
wget -q https://downloads.sourceforge.net/glew/glew-2.2.0-win32.zip
unzip -q glew-2.2.0-win32.zip
mv glew-2.2.0 glew
rm glew-2.2.0-win32.zip


echo "=== Fixing libogg config_types.h ==="

cp libogg/config_types.h libogg/include/ogg/ 2>/dev/null || true


echo "=== Creating symlinks ==="

cd ~/RSDKv3-Decompilation/android/app/jni

rm -f SDL src

ln -s ../../../dependencies/android/SDL SDL
ln -s ../../../.. src


echo "=== Checking files ==="

test -f ../../../dependencies/android/libtheora/lib/analyze.c
test -f ../../../dependencies/android/libtheora/lib/cpu.c

echo "Dependencies installed successfully!"
```

```python
%%bash

cd ~/RSDKv3-Decompilation/android

chmod +x gradlew

export GRADLE_OPTS="-Xmx4g"

./gradlew assembleDebug
```
